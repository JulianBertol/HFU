Importieren der CSV Datei und ausgeben

```python
import csv

unternehmens_liste = []
with open("50Data.csv", newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=',')
    for row in reader:
        unternehmen = row['Unternehmen']
        kaufmenge = int(row['Kaufmenge'])
        unternehmens_liste.append((unternehmen, kaufmenge))

print(unternehmens_liste)
```

Mittelwert bestimmen

```python
def get_avg(data):
  sum = 0
  for line in data:
      sum += line
  avg = sum / len(data)
  return avg

kaufmengen = [kaufmenge for _, kaufmenge in unternehmens_liste]
get_avg(kaufmengen)
```

Standardabweichung berechnen

```python
import math
def standard_deviation(data):
  standard_deviation = 0
  denominator = 0
  avg = get_avg(data)
  for line in data:
    denominator += pow(line - avg, 2)
  standard_deviation = (denominator / (len(data) - 1))
  standard_deviation = math.sqrt (standard_deviation)
  return standard_deviation

kaufmengen = [kaufmenge for _, kaufmenge in unternehmens_liste]
standard_deviation(kaufmengen)
```

importieren der 1000Data csv

```python
import csv
import math
import random

Air_temperature = []
relative_humidity = []
stations = []

with open("1000Data.csv", newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=',')
    for row in reader:
        Air_temperature.append(float(row[' LUFTTEMPERATUR']))
        relative_humidity.append(float(row[' REL_FEUCHTE']))
        stations.append(row['STATIONS_ID'])  # Station mitladen
```

Berechnen der Mittelwerte mit entsprechenden Rauschwerten

```python
def apply_noise(data, max_percent):
    """Wendet zufälliges Rauschen (0 bis max_percent%) auf alle Werte an."""
    noisy_data = []
    for value in data:
        percent = random.uniform(0, max_percent / 100)  # z.B. 0 bis 0.05
        sign = random.choice([-1, 1])
        noise = value * percent
        noisy_data.append(value + sign * noise)
    return noisy_data

def get_avg(data):
    total = 0
    for value in data:
        total += value
    return total / len(data)

def standard_deviation(data, avg):
    denominator = 0
    for line in data:
        denominator += pow(line - avg, 2)
    return math.sqrt(denominator / (len(data) - 1))


def save_to_csv(filename, temp_data, hum_data, avg_temp, std_temp, avg_hum, std_hum):
    with open(filename, "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        # Header
        writer.writerow([
            "LUFTTEMPERATUR", "REL_FEUCHTE",
            "Avg_Lufttemperatur", "Std_Lufttemperatur",
            "Avg_Rel_Feuchte", "Std_Rel_Feuchte"
        ])
        # Werte + Statistik in erster Zeile
        for i in range(len(temp_data)):
            if i == 0:
                writer.writerow([
                    temp_data[i], hum_data[i],
                    avg_temp, std_temp,
                    avg_hum, std_hum
                ])
            else:
                writer.writerow([temp_data[i], hum_data[i], "", "", "", ""])
```

```python
experiments = [
    (0,  "0Prozent-Rausch-Data.csv"),
    (5,  "5Prozent-Rausch-Data.csv"),
    (10, "10Prozent-Rausch-Data.csv"),
    (50, "50Prozent-Rausch-Data.csv"),
]

for max_percent, filename in experiments:
    # Rauschen anwenden (bei 0% bleibt alles gleich)
    noisy_temp = apply_noise(Air_temperature, max_percent)
    noisy_hum  = apply_noise(relative_humidity, max_percent)

    # Mittelwert & Standardabweichung berechnen
    avg_temp = get_avg(noisy_temp)
    std_temp = standard_deviation(noisy_temp, avg_temp)
    avg_hum  = get_avg(noisy_hum)
    std_hum  = standard_deviation(noisy_hum, avg_hum)

    # CSV speichern
    save_to_csv(filename, noisy_temp, noisy_hum, avg_temp, std_temp, avg_hum, std_hum)

    # Ausgabe in Konsole
    print(f"{max_percent}% Rauschen ({filename})")
    print(f"  Lufttemperatur Mittelwert: {avg_temp:.4f}, Standardabweichung: {std_temp:.4f}")
    print(f"  Relative Feuchte Mittelwert: {avg_hum:.4f}, Standardabweichung: {std_hum:.4f}\n")
```
