# Aufgabe 1

```python
import pandas as pd

df = pd.read_csv('AirQualityMissingData.csv')
print(len(df))
```

# Aufgabe 2

```python
# Die 4 relevanten Spalten auswählen
spalten = df[['Ozone', 'Solar', 'Wind', 'Temp']]

# Für jede Zelle prüfen ob sie leer ist
ist_leer = spalten.isnull()

# Die True-Werte pro Spalte zusammenzählen
anzahl_leer = ist_leer.sum()

# Ergebnis ausgeben
print("Anzahl fehlender Werte pro Spalte:")
print(anzahl_leer)
```

# Aufgabe 3

```python
# Alle Zeilen löschen, die mindestens einen leeren Wert enthalten
df_dropped = df.dropna()

# Vorher / Nachher vergleichen
print("Vorher:", len(df))       # 153
print("Nachher:", len(df_dropped))  # 111
print("Gelöscht:", len(df) - len(df_dropped))  # 42
```

## Grafen Zeichnen

```python
import matplotlib.pyplot as plt

# 4 Subplots erstellen (2 Zeilen, 2 Spalten)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Nach Listwise Deletion (111 Datenvektoren)', fontsize=14, fontweight='bold')

# Ozone
axes[0, 0].plot(df_dropped['Ozone'], color='steelblue')
axes[0, 0].set_title('Ozone')
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Wert')

# Solar
axes[0, 1].plot(df_dropped['Solar'], color='darkorange')
axes[0, 1].set_title('Solar')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Wert')

# Wind
axes[1, 0].plot(df_dropped['Wind'], color='seagreen')
axes[1, 0].set_title('Wind')
axes[1, 0].set_xlabel('Index')
axes[1, 0].set_ylabel('Wert')

# Temp
axes[1, 1].plot(df_dropped['Temp'], color='crimson')
axes[1, 1].set_title('Temp')
axes[1, 1].set_xlabel('Index')
axes[1, 1].set_ylabel('Wert')

plt.tight_layout()
plt.show()
```

# Aufgabe 4

```python
# Kopie erstellen
df_mean = df.copy()

# Jede Spalte: leere Werte mit dem Mittelwert der Spalte füllen
df_mean['Ozone'] = df_mean['Ozone'].fillna(df_mean['Ozone'].mean())
df_mean['Solar'] = df_mean['Solar'].fillna(df_mean['Solar'].mean())
df_mean['Wind']  = df_mean['Wind'].fillna(df_mean['Wind'].mean())
df_mean['Temp']  = df_mean['Temp'].fillna(df_mean['Temp'].mean())

# Graph
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Methode: Mean Value Imputation', fontsize=14, fontweight='bold')

axes[0, 0].plot(df_mean['Ozone'], color='steelblue')
axes[0, 0].set_title('Ozone')

axes[0, 1].plot(df_mean['Solar'], color='darkorange')
axes[0, 1].set_title('Solar')

axes[1, 0].plot(df_mean['Wind'], color='seagreen')
axes[1, 0].set_title('Wind')

axes[1, 1].plot(df_mean['Temp'], color='crimson')
axes[1, 1].set_title('Temp')

plt.tight_layout()
plt.show()
```

# Aufgabe 5

```python
# Kopie erstellen
df_median = df.copy()

# Leere Werte mit dem Median der Spalte füllen
df_median['Ozone'] = df_median['Ozone'].fillna(df_median['Ozone'].median())
df_median['Solar'] = df_median['Solar'].fillna(df_median['Solar'].median())
df_median['Wind']  = df_median['Wind'].fillna(df_median['Wind'].median())
df_median['Temp']  = df_median['Temp'].fillna(df_median['Temp'].median())

# Graph
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Methode: Median Value Imputation', fontsize=14, fontweight='bold')

axes[0, 0].plot(df_median['Ozone'], color='steelblue')
axes[0, 0].set_title('Ozone')

axes[0, 1].plot(df_median['Solar'], color='darkorange')
axes[0, 1].set_title('Solar')

axes[1, 0].plot(df_median['Wind'], color='seagreen')
axes[1, 0].set_title('Wind')

axes[1, 1].plot(df_median['Temp'], color='crimson')
axes[1, 1].set_title('Temp')

plt.tight_layout()
plt.show()
```

# Aufgabe 6

```python
# Kopie erstellen
df_locf = df.copy()

# ffill = "forward fill" = letzten bekannten Wert vorwärts übertragen
df_locf['Ozone'] = df_locf['Ozone'].ffill()
df_locf['Solar'] = df_locf['Solar'].ffill()
df_locf['Wind']  = df_locf['Wind'].ffill()
df_locf['Temp']  = df_locf['Temp'].ffill()

# Graph
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Methode: LOCF (Last Observation Carried Forward)', fontsize=14, fontweight='bold')

axes[0, 0].plot(df_locf['Ozone'], color='steelblue')
axes[0, 0].set_title('Ozone')

axes[0, 1].plot(df_locf['Solar'], color='darkorange')
axes[0, 1].set_title('Solar')

axes[1, 0].plot(df_locf['Wind'], color='seagreen')
axes[1, 0].set_title('Wind')

axes[1, 1].plot(df_locf['Temp'], color='crimson')
axes[1, 1].set_title('Temp')

plt.tight_layout()
plt.show()
```

# Aufgabe 7

```python
# Kopie erstellen
df_nocb = df.copy()

# bfill = "backward fill" = nächsten bekannten Wert rückwärts übertragen
df_nocb['Ozone'] = df_nocb['Ozone'].bfill()
df_nocb['Solar'] = df_nocb['Solar'].bfill()
df_nocb['Wind']  = df_nocb['Wind'].bfill()
df_nocb['Temp']  = df_nocb['Temp'].bfill()

# Graph
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Methode: NOCB (Next Observation Carried Backward)', fontsize=14, fontweight='bold')

axes[0, 0].plot(df_nocb['Ozone'], color='steelblue')
axes[0, 0].set_title('Ozone')

axes[0, 1].plot(df_nocb['Solar'], color='darkorange')
axes[0, 1].set_title('Solar')

axes[1, 0].plot(df_nocb['Wind'], color='seagreen')
axes[1, 0].set_title('Wind')

axes[1, 1].plot(df_nocb['Temp'], color='crimson')
axes[1, 1].set_title('Temp')

plt.tight_layout()
plt.show()
```

# Aufgabe 8

```python
# Kopie erstellen
df_interp = df.copy()

# Lineare Interpolation
df_interp['Ozone'] = df_interp['Ozone'].interpolate(method='linear')
df_interp['Solar'] = df_interp['Solar'].interpolate(method='linear')
df_interp['Wind']  = df_interp['Wind'].interpolate(method='linear')
df_interp['Temp']  = df_interp['Temp'].interpolate(method='linear')

# Falls NaN am Anfang (kein Vorgänger vorhanden) → mit bfill auffüllen
df_interp['Ozone'] = df_interp['Ozone'].bfill()
df_interp['Solar'] = df_interp['Solar'].bfill()

# Graph
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Methode: Lineare Interpolation', fontsize=14, fontweight='bold')

axes[0, 0].plot(df_interp['Ozone'], color='steelblue')
axes[0, 0].set_title('Ozone')

axes[0, 1].plot(df_interp['Solar'], color='darkorange')
axes[0, 1].set_title('Solar')

axes[1, 0].plot(df_interp['Wind'], color='seagreen')
axes[1, 0].set_title('Wind')

axes[1, 1].plot(df_interp['Temp'], color='crimson')
axes[1, 1].set_title('Temp')

plt.tight_layout()
plt.show()
```

## PCAP Datei Laden und ausgeben

```python
!pip install dpkt networkx

import dpkt
import collections

def parse_pcap(filepath):
    protocols = collections.Counter()
    ip_traffic = collections.Counter()
    edges = collections.Counter()

    with open(filepath, 'rb') as f:
        cap = dpkt.pcapng.Reader(f)

        for ts, buf in cap:
            try:
                eth = dpkt.ethernet.Ethernet(buf)
            except Exception:
                continue

            if isinstance(eth.data, dpkt.ip.IP):
                ip = eth.data
                src = dpkt.utils.inet_to_str(ip.src)
                dst = dpkt.utils.inet_to_str(ip.dst)

                # IP-Adressen zählen
                ip_traffic[src] += 1
                ip_traffic[dst] += 1

                # Verbindungen zählen
                edges[(src, dst)] += 1

                # Protokoll bestimmen
                if isinstance(ip.data, dpkt.tcp.TCP):
                    protocols['TCP'] += 1
                elif isinstance(ip.data, dpkt.udp.UDP):
                    protocols['UDP'] += 1
                elif isinstance(ip.data, dpkt.icmp.ICMP):
                    protocols['ICMP'] += 1
                else:
                    protocols['IP Other'] += 1
            else:
                protocols['Non-IP'] += 1

    return protocols, ip_traffic, edges

# Datei laden
protocols, ip_traffic, edges = parse_pcap('demo.pcapng')

print("Protokolle:", dict(protocols))
print("IP-Adressen:", dict(ip_traffic))
print("Verbindungen:", dict(edges))
```

## IP Adressen als Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# Gerichteten Graphen erstellen
G = nx.DiGraph()

# Knoten hinzufügen (die IP-Adressen)
G.add_node('192.168.3.65')
G.add_node('188.72.243.72')

# Kanten hinzufügen (Verbindungen mit Paketanzahl)
G.add_edge('192.168.3.65', '188.72.243.72', weight=425)
G.add_edge('188.72.243.72', '192.168.3.65', weight=680)

# Position der Knoten festlegen
pos = nx.spring_layout(G, seed=42)

# Graph zeichnen
fig, ax = plt.subplots(figsize=(8, 6))

nx.draw_networkx_nodes(G, pos, node_size=3000,
                       node_color='#ADD8E6', ax=ax)
nx.draw_networkx_labels(G, pos, font_size=9,
                        font_weight='bold', ax=ax)
nx.draw_networkx_edges(G, pos, width=2,
                       edge_color='#555555',
                       arrows=True, arrowsize=20,
                       connectionstyle='arc3,rad=0.2',
                       ax=ax)

edge_labels = {('192.168.3.65', '188.72.243.72'): '425 Pakete',
               ('188.72.243.72', '192.168.3.65'): '680 Pakete'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels,
                              font_size=9, ax=ax)

ax.set_title('IP Connection Graph', fontsize=14, fontweight='bold')
ax.axis('off')
plt.tight_layout()
plt.show()
```

## nochmal mit bereinigten IP Adressen

```python
import socket
import struct

# IP-Adresse in Integer umwandeln
def ip_zu_int(ip):
    return struct.unpack('!I', socket.inet_aton(ip))[0]

# IPs normalisieren (0 bis 1)
def normalisieren(ip_liste):
    int_werte = {ip: ip_zu_int(ip) for ip in ip_liste}
    minimum = min(int_werte.values())
    maximum = max(int_werte.values())
    return {ip: (wert - minimum) / (maximum - minimum)
            for ip, wert in int_werte.items()}

# Unsere IP-Adressen normalisieren
ip_liste = ['192.168.3.65', '188.72.243.72']
norm = normalisieren(ip_liste)

print("Normalisierte Werte:")
for ip, wert in norm.items():
    print(f"  {ip} → {wert:.4f}")

# Knotennamen mit normalisierten Werten
label_192 = f"192.168.3.65\n(norm: {norm['192.168.3.65']:.4f})"
label_188 = f"188.72.243.72\n(norm: {norm['188.72.243.72']:.4f})"

# Graph erstellen
G2 = nx.DiGraph()
G2.add_node(label_192)
G2.add_node(label_188)
G2.add_edge(label_192, label_188, weight=425)
G2.add_edge(label_188, label_192, weight=680)

# Graph zeichnen
pos2 = nx.spring_layout(G2, seed=42)
fig, ax = plt.subplots(figsize=(10, 6))

nx.draw_networkx_nodes(G2, pos2, node_size=4000,
                       node_color='#ADD8E6', ax=ax)
nx.draw_networkx_labels(G2, pos2, font_size=8,
                        font_weight='bold', ax=ax)
nx.draw_networkx_edges(G2, pos2, width=2,
                       edge_color='#555555',
                       arrows=True, arrowsize=20,
                       connectionstyle='arc3,rad=0.2',
                       ax=ax)

edge_labels2 = {(label_192, label_188): '425 Pakete',
                (label_188, label_192): '680 Pakete'}
nx.draw_networkx_edge_labels(G2, pos2, edge_labels=edge_labels2,
                              font_size=8, ax=ax)

ax.set_title('IP Connection Graph (Normalisiert)', fontsize=14, fontweight='bold')
ax.axis('off')
plt.tight_layout()
plt.show()
```
